## Introducción:
El tiempo dedicado por los empleados al desarrollo de las actividades dentro de la empresa resulta muy valioso para la toma de decisiones. Esta información se recaba mediante una 
aplicación donde cada trabajador debe medir el tiempo que le dedica a sus tareas diarias, por este medio se recaba la información la cual resulta muy útil ante la toma de decisiones tanto 
estrátegicas como operativas dentro de la misma. El período de análisis del siguiente informe es del primer trimestre del 2022. Al contar con estos datos surgen varias preguntan a contesta como, 
desde el lado del servicio que se brinda por la empresa *Cuales categorías (servicios) captan la mayor cantidad de horas por el área de diseño?

Cuál es la saturación por empleado que se observa en el primer trimestre del año?

```python
import pandas as pd
```

#### Se investiga los datos del dataframe y luego se procede a limpiar la base de datos para poder obtener la información necesaria mediante herramientas como drop y rename

```python
hrs.head()
```

```python
hrs.columns
```

```python
labels = ['Proyecto', 'Tarea', 'Notas', 'Usuario', 'Fecha', 'Desde', 'hasta',
       'Duración', 'Horas', 'Social Media', 'Performance', 'Diseño', 'Growth']   
```

```python
columns ={}

for label in labels:
    
    columns[label] = label.replace(' ', '_').lower()
                                                             #transformación de los encabezados según las buenas practicas de Python
                                                                            
```

```python
hrs = hrs.rename(columns=columns)
hrs
```

```python
hrs = hrs.drop(['social_media','growth','performance','duración'], axis = 1)                #se eliminan las columnas que no son de interés

hrs
```

```python
hrs = hrs.rename(columns = {'proyecto':'clientes','notas':'reuniones',
                            'usuario':'empleado','diseño':'cambios'})         

                                                                                                #se renombrea columnas para el análisis
hrs
```

```python
hrs['fecha'] = pd.to_datetime(hrs['fecha'], dayfirst=True)

hrs.fecha
```

```python
index_name = hrs [(hrs['horas'] >8)].index
hrs.drop(index_name, inplace=True)
 
hrs                                                              #El horario laboral es de 8 hrs diarias, se eliminan tareas de mayor duración de hrs que 
                                                                #pueden haberse olvidado de parar reloj.
```

```python
hrs.horas.mean()
```

```python
import numpy as np
import datetime
```

```python
enero = '2022-01-01'
febrero = '2022-02-01'
marzo = '2022-03-01'
abril = '2022-04-01' 
```

```python
mask = (hrs['fecha'] >= enero ) & (hrs['fecha'] < febrero)
q1_enero = hrs.loc[mask]
q1_enero
```

```python
mask = (hrs['fecha'] >= febrero ) & (hrs['fecha'] < marzo)
q1_febrero = hrs.loc[mask]
q1_febrero
```

```python
mask = (hrs['fecha'] >= marzo) & (hrs['fecha'] < abril)
q1_marzo = hrs.loc[mask]
q1_marzo
```

```python
hrs_enero = q1_enero['horas'].sum()
hrs_febrero = q1_febrero['horas'].sum()
hrs_marzo = q1_marzo['horas'].sum()

print(f'las horas totales trackeadas en el primer trimestre son: {hrs_enero + hrs_febrero + hrs_marzo}') 

                                     
print(f'la cantidad de horas trackeadas en marzo ({round(hrs_marzo,2)}) es mayor con respecto a las de enero ({hrs_enero}) y febrero ({hrs_febrero})')                                                
```

```python
cat_media = hrs['tarea'].str.contains('MEDIA', case=False, na=False)
media = hrs.loc[cat_media]
media
```

```python
hrs_ads_q1 = ads['horas'].sum()
hrs_ads_q1 
```

```python
cat_web = hrs['tarea'].str.contains('WEB', case=False, na=False)
web = hrs.loc[cat_web]
web
```

```python
hrs_web_q1 = web['horas'].sum()
hrs_web_q1 
```

```python
cat_tradicional = hrs['tarea'].str.contains('TRADICIONAL', case=False, na=False)
tradicional = hrs.loc[cat_tradicional]
tradicional
```

```python
hrs_tradicional_q1 = tradicional['horas'].sum()
hrs_tradicional_q1 
```

```python
cat_audiovisual = hrs['tarea'].str.contains('AUDIOVISUAL', case=False, na=False)
audiovisual = hrs.loc[cat_audiovisual]
audiovisual
```

### Saturación Empleado
* Enero: 140
* Febrero: 133
* Marzo: 154

* Total de horas por empleado en Q1: 427 con un margen de error de una hora por día. Por lo que se observa que Sofia se encuentra cerca del nivel de tope de trabajo diario, tiene sentido dado que es la head del área y además de diseñar desarrolla actividades de liderazgo en el área. Jerusa y Gonzalo muestran un margen considerable de tiempo libre para poder realizar actividades que les compete al igual que Matias, quien se encuentra muy por debajo de la media de las horas de trabajo.

```python
sofia_q1 = (hrs['empleado'].str.contains('Sofia', case=False))
sofia = hrs.loc[sofia_q1]
hrs_sofia = sofia['horas'].sum()

gonzalo_q1 = (hrs['empleado'].str.contains('gonzalo', case=False))
gonzalo = hrs.loc[gonzalo_q1]
hrs_gonzalo = gonzalo['horas'].sum()

jerusa_q1 = (hrs['empleado'].str.contains('jerusa', case=False))
jerusa = hrs.loc[jerusa_q1]
hrs_jerusa = jerusa['horas'].sum()

matias_q1 = (hrs['empleado'].str.contains('Matias', case=False))
matias = hrs.loc[matias_q1]
hrs_matias = matias['horas'].sum()
```

```python
print(f'La cantidad de horas medidas por empleado son:')
      
print(f'* Sofia Pérez:{round(hrs_sofia,2)}, Gonzalo Zamora:{round(hrs_gonzalo,2)}, Jerusa Villalba:{round(hrs_jerusa,2)}, Matias Paoli:{round(hrs_matias,2)}')      
      


```
